```python
import pandas as pd
import numpy as np

# Define options for categorical features
cuts = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
colors = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarities = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']

# Set a seed for reproducibility
np.random.seed(42)

# Number of samples
n = 10

# Generate data
data = {
    'carat': np.round(np.random.uniform(0.2, 2.5, n), 2),
    'cut': np.random.choice(cuts, n),
    'color': np.random.choice(colors, n),
    'clarity': np.random.choice(clarities, n),
    'depth': np.round(np.random.uniform(55, 65, n), 1),
    'table': np.round(np.random.uniform(50, 70, n), 1),
    'price': np.random.randint(300, 20000, n),
    'x': np.round(np.random.uniform(3.5, 9.5, n), 2),
    'y': np.round(np.random.uniform(3.5, 9.5, n), 2),
    'z': np.round(np.random.uniform(2.0, 6.0, n), 2)
}

# Create DataFrame
df = pd.DataFrame(data)
df.index.name = 'id'

print(df)
```

```python
sns.histplot(df['carat'], kde=True, bins=10)
plt.title("Distribution of Diamond Carat")
plt.xlabel("carat")
plt.ylabel("colors")
plt.show()
```

```python
sns.jointplot(data=df, x="carat", y="price", hue="cut", kind="scatter")
plt.show()
```

```python


data = {
    'customer_id': [f'CUST{1000 + i}' for i in range(n)],
    'gender': np.random.choice(['Male', 'Female'], n),
    'age': np.random.randint(18, 70, n),
    'tenure_months': np.random.randint(1, 72, n),
    'service_plan': np.random.choice(['Basic', 'Standard', 'Premium'], n),
    'monthly_charges': np.round(np.random.uniform(20.0, 120.0, n), 2),
    'internet_service': np.random.choice(['DSL', 'Fiber optic', 'None'], n),
    'tech_support': np.random.choice(['Yes', 'No'], n),
    'contract_type': np.random.choice(['Month-to-month', 'One year', 'Two year'], n),
    'payment_method': np.random.choice(['Credit card', 'Bank transfer', 'Electronic check', 'Mailed check'], n),
    'churn': np.random.choice([0, 1], n, p=[0.7, 0.3])  # 0 = No churn, 1 = Churned
}

df=pd.DataFrame(data)

df


```

```python
sns.violinplot(data=df, x='service_plan', y='monthly_charges', hue='churn', split=True)
plt.title('Monthly Charges by Service Plan and Churn')
plt.xlabel('Service Plan')
plt.ylabel('Monthly Charges')
plt.show()
```

```python
sns.boxplot(data=df,x='service_plan',y='monthly_charges',hue='churn')

plt.title('Monthly charges by service plan')

plt.xlabel('Service Plan')

plt.ylabel('Monthly charges')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.boxplot(data=df, x='gender', y='age', ax=axes[0])
axes[0].set_title("Boxplot of Age by Gender")

sns.violinplot(data=df, x='gender', y='age', hue='churn', split=True, ax=axes[1])
axes[1].set_title("Violinplot of Age by Gender and Churn")

plt.tight_layout()
plt.show()
```
