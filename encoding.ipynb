```python
import numpy as npt
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as pl
```

```python
df=pd.read_csv('Titanic-Dataset.csv')

df
```

```python
from sklearn.preprocessing import OneHotEncoder

#Display initial categorical columns

cat_cols=df.select_dtypes(include='object')

df_copy=cat_cols[['Sex']].copy()

#Initialize and fit onehotencoder

enc=OneHotEncoder(sparse_output=False)
encoded_array=enc.fit_transform(df_copy)

#Get the eencoded category names
encoded_feature_names=enc.get_feature_names_out(['Sex'])

#Convert encoded array to dataframe with column names

encoded_df=pd.DataFrame(encoded_array,columns=encoded_feature_names)

#Concatenate original and encoded columns

df_final=pd.concat([df_copy,encoded_df],axis=1)

#Result

df_final.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder=OrdinalEncoder(categories=[['female','male']])
df['Sex_encoded']=encoder.fit_transform(df[['Sex']])
print("\nEncoded Data:")
print(df[['Sex','Sex_encoded']].head())
```
