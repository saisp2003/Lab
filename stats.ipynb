```python
for column in df.columns:
    plt.figure(figsize=(6,4))
    sms.histplot(df[column],kde=True,color='red')
    plt.title(f'Gaussian Distribution of{column}')
    plt.xlabel('column')
    plt.ylabel('Density')
    plt.grid(False)
    plt.show()
```

```python
# Calculate Q1 and Q3
Q1 = df.quantile(0.25,numeric_only=True)
Q3 = df.quantile(0.75,numeric_only=True)

# Calculate IQR
IQR = Q3 - Q1

print("Interquartile Range (IQR):")
print(IQR)
```

```python
#Create heatmap

plt.Figure(figsize=(8,6))
sms.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('correlation heatmap of iris dataset')
plt.show()
```

```python
#calculate skewness

skew_values=df.skew(numeric_only=True)
print('Skewness of each feature:')
print(skew_values)

#plot the distributions to visualise skewness

for column in df.columns:
    sms.histplot(df[column],kde=True)
    plt.title(f'Distribution and skewness of{column}')
    plt.xlabel('Column')
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sms.barplot(x='species',y='sepalLength',data=df,estimator=np.mean)
plt.title('Mean Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average sepal length')
plt.show()
```

```python
#Count Plot

plt.figure(figsize=(6,4))
sms.countplot(x='species',data=df)
plt.title('Count Plot of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Pie Chart

species_count = df['species'].value_counts()

plt.figure(figsize=(6,4))
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=120)
plt.title("Distribution of Species")
plt.axis('equal')
plt.show()
```

```python
#add a new categorical column with 3 flower types

def assign_flower_type(species):
    if species=='setosa':
        return 'Type A'
    elif species=='versicolor':
        return'Type B'
    else:
        return 'Type C'

df['flower_type']=df['species'].apply(assign_flower_type)

contigency_table=pd.crosstab(df['species'],df['flower_type'])
print(contigency_table)
```
